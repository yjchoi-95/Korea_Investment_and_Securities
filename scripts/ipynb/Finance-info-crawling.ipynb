```python
import pandas_datareader as pdr
import yfinance as yf
```

```python
cnt = 0
```

```python
for idx, code in tqdm(enumerate(tot_df.Symbol)):
    try :
        temp_df  = pdr.data.get_quote_yahoo(code)
    except :
        continue
        
    if cnt == 0:
        info_df = temp_df
        cnt += 1
    else :
        info_df = info_df.append(temp_df)
```

```python
info_df.to_csv("info_df.csv")
```
